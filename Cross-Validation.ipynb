# Cross-Validation

```python
from sklearn.datasets import load_iris
```

```python
iris_data=load_iris()
print(iris_data)
```

```python
data_input=iris_data.data     #indepedent
data_output=iris_data.target  # dependent
```

```python
data_input
```

```python
data_output
```

```python
data_output.shape
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5, shuffle=False)
```

```python
print("Train set              Test set           ")
for train_set,test_set in kf.split(data_input):
    print(train_set,test_set)
```

```python
kf=KFold(n_splits=10, shuffle=True)

print("Train set              Test set           ")
for train_set,test_set in kf.split(data_input):
    print(train_set,test_set)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_class=RandomForestClassifier(n_estimators=10)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(rf_class, data_input, data_output, scoring='accuracy',cv=5))  # val of k
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(rf_class, data_input, data_output, scoring='accuracy',cv=5).mean()) 
```
